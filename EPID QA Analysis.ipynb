## Main Function

```python
def display_results(data, comData = [], stats=False, graph=True, commissioning=False, title1 = "", title2 = ""):

    if graph:
        fig = plt.figure(figsize=(20,5))
        ax1 = fig.add_subplot(121)
        ax1.set_title(title1)
        ax2 = fig.add_subplot(122)
        ax2.set_title(title2)
    x = data


    for col in x.columns: 
        #print(col)
        #print(x[col].dropna())
        new = x[col].apply(pd.to_numeric, errors='coerce')
        new = new.dropna()
        if graph:
            new.plot(ax=ax1)
                
        new = new[new.between(new.quantile(.05), new.quantile(.95))] # without outliers
        #print(new)
        if graph:
            new.plot(ax=ax2)
            if commissioning:
                plt.axhline(y=comData)
            plt.legend()
        if stats:
            print(col," Mean: ",round(new.mean(),2))
            print("    Min: ",round(new.min(),2))
            print("    Max: ",round(new.max(),2))
            print()
            
            
# Setup Random Noise
# Setup CNR
# Setup MTF
# HDR Random Noise
# HDR CNR
# HDR MTF
# LDR Random Noise
# LDR CNR
# LDR MTF
# Matlab Measured Graticule Length
# Offline Review Length
# Gantry 0 Image Centre
# Gantry 90 Image Centre
# Gantry 180 Image Centre
# Gantry 270 Image Centre
# L/R 6MV
# Vertical Field Size
# Horizontal Field Size
```
