```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
sns.set()
```

```python
def makeBar(iroc2020):
    fig, axes = plt.subplots(3,3, sharey=True, figsize=(25,18))

    #unit 1
    DATA = iroc2020.loc[iroc2020['Unit'] == 1.0]
    ax = DATA.plot.bar(x='Beam',y=["IROC Houston Measured Dose at dmax (cGy)","Instituion Reported Dose at dmax (cGy)"], ax = axes[0,0])
    plt.ylim([98,106])
    axes[0,0].set_ylabel("%")
    axes[0,0].set_title("Unit 1")


    for p in ax.patches:
        ax.annotate("%.2f" % p.get_height(), (p.get_x() + p.get_width() / 2., p.get_height()), 
                    ha='center', va='center', xytext=(0, 10), textcoords='offset points',fontsize=15)


    #unit 2
    DATA = iroc2020.loc[iroc2020['Unit'] == 2.0]
    ax = DATA.plot.bar(x='Beam',y=["IROC Houston Measured Dose at dmax (cGy)","Instituion Reported Dose at dmax (cGy)"], ax = axes[0,1])
    plt.ylim([98,106])
    axes[0,1].set_title("Unit 2")

    for p in ax.patches:
        ax.annotate("%.2f" % p.get_height(), (p.get_x() + p.get_width() / 2., p.get_height()), 
                    ha='center', va='center', xytext=(0, 10), textcoords='offset points',fontsize=15)

    #unit 3
    DATA = iroc2020.loc[iroc2020['Unit'] == 3.0]
    ax = DATA.plot.bar(x='Beam',y=["IROC Houston Measured Dose at dmax (cGy)","Instituion Reported Dose at dmax (cGy)"], ax = axes[0,2])
    plt.ylim([98,106])
    axes[0,2].set_title("Unit 3")

    for p in ax.patches:
        ax.annotate("%.2f" % p.get_height(), (p.get_x() + p.get_width() / 2., p.get_height()), 
                    ha='center', va='center', xytext=(0, 10), textcoords='offset points',fontsize=15)

    #unit 5
    DATA = iroc2020.loc[iroc2020['Unit'] == 5.0]
    ax = DATA.plot.bar(x='Beam',y=["IROC Houston Measured Dose at dmax (cGy)","Instituion Reported Dose at dmax (cGy)"], ax = axes[1,0])
    plt.ylim([98,106])
    axes[1,0].set_title("Unit 5")
    axes[1,0].set_ylabel("%")

    for p in ax.patches:
        ax.annotate("%.2f" % p.get_height(), (p.get_x() + p.get_width() / 2., p.get_height()), 
                    ha='center', va='center', xytext=(0, 10), textcoords='offset points',fontsize=15)

    #unit 6
    DATA = iroc2020.loc[iroc2020['Unit'] == 6.0]
    ax = DATA.plot.bar(x='Beam',y=["IROC Houston Measured Dose at dmax (cGy)","Instituion Reported Dose at dmax (cGy)"], ax = axes[1,1])
    plt.ylim([98,106])
    axes[1,1].set_title("Unit 6")

    for p in ax.patches:
        ax.annotate("%.2f" % p.get_height(), (p.get_x() + p.get_width() / 2., p.get_height()), 
                    ha='center', va='center', xytext=(0, 10), textcoords='offset points',fontsize=15)

    #unit 7
    DATA = iroc2020.loc[iroc2020['Unit'] == 7.0]
    ax = DATA.plot.bar(x='Beam',y=["IROC Houston Measured Dose at dmax (cGy)","Instituion Reported Dose at dmax (cGy)"], ax = axes[1,2])
    plt.ylim([98,106])
    axes[1,2].set_title("Unit 7")

    for p in ax.patches:
        ax.annotate("%.2f" % p.get_height(), (p.get_x() + p.get_width() / 2., p.get_height()), 
                    ha='center', va='center', xytext=(0, 10), textcoords='offset points',fontsize=15)

    #unit 8
    DATA = iroc2020.loc[iroc2020['Unit'] == 8.0]
    ax = DATA.plot.bar(x='Beam',y=["IROC Houston Measured Dose at dmax (cGy)","Instituion Reported Dose at dmax (cGy)"], ax = axes[2,0])
    plt.ylim([98,106])
    axes[2,0].set_title("Unit 8")
    axes[2,0].set_ylabel("%")

    for p in ax.patches:
        ax.annotate("%.2f" % p.get_height(), (p.get_x() + p.get_width() / 2., p.get_height()), 
                    ha='center', va='center', xytext=(0, 10), textcoords='offset points',fontsize=15)

    #unit 9
    DATA = iroc2020.loc[iroc2020['Unit'] == 9.0]
    ax = DATA.plot.bar(x='Beam',y=["IROC Houston Measured Dose at dmax (cGy)","Instituion Reported Dose at dmax (cGy)"], ax = axes[2,1])
    plt.ylim([98,106])
    axes[2,1].set_title("Unit 9")

    for p in ax.patches:
        ax.annotate("%.2f" % p.get_height(), (p.get_x() + p.get_width() / 2., p.get_height()), 
                    ha='center', va='center', xytext=(0, 10), textcoords='offset points',fontsize=15)

    #unit 10
    DATA = iroc2020.loc[iroc2020['Unit'] == 10.0]
    ax = DATA.plot.bar(x='Beam',y=["IROC Houston Measured Dose at dmax (cGy)","Instituion Reported Dose at dmax (cGy)"], ax = axes[2,2])
    plt.ylim([98,106])
    axes[2,2].set_title("Unit 10")

    for p in ax.patches:
        ax.annotate("%.2f" % p.get_height(), (p.get_x() + p.get_width() / 2., p.get_height()), 
                    ha='center', va='center', xytext=(0, 10), textcoords='offset points',fontsize=15)

    plt.tight_layout()
```

# Heatmap

```python
def heatmap():
    sns.set(style="white")
    iroc2020 = pd.read_csv("iroc2020.csv")
    iroc2020 = iroc2020.dropna()
    iroc2020_percentages = iroc2020[["Unit","Beam","PD"]]
    iroc2020_percentages["Unit"] = iroc2020["Unit"].astype(int) 

    cmap = sns.diverging_palette(220, 10, as_cmap=True)

    # Draw a heatmap with the numeric values in each cell
    f, ax = plt.subplots(figsize=(9, 6))
    iroc2020_percentages2 = iroc2020_percentages.pivot("Unit", "Beam", "PD")
    sns.heatmap(iroc2020_percentages2, annot=True, linewidths=.5, ax=ax, cmap=cmap, cbar_kws={'label': 'Absolute Percent Difference (%)'},vmin=0, vmax=5)
```

```python
heatmap()
```
